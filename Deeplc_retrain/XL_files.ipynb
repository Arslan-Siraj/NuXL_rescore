```python
""" read idXML file and converts to desired dataframes """
def Conversion_to_DF(file_name):
    full_file_name = file_name + ".mzTab"
    file1 = open(full_file_name, 'r')
    Lines = file1.readlines()

    lines = len(Lines)
    indices = [index for index, element in enumerate(Lines) if element == "\n"]

    #Get the chuncks 1: Description, 2: Proteins, 3: PSMs, Seperated
    MetaValue = Lines[0:indices[0]]
    PSMs = Lines[indices[0]+1:]

    MetaValue_sp = []
    for i in MetaValue:
        x = i.split('\t')
        MetaValue_sp.append(x)
    Desc_df = pd.DataFrame(MetaValue_sp)

    PSMs_sp = []
    for i in PSMs:
        x = i.split('\t')
        PSMs_sp.append(x)
    PSMs_df_ = pd.DataFrame(PSMs_sp)


    #Select first row as header
    PSMs_df_.columns = PSMs_df_.iloc[0]
    PSMs_df= PSMs_df_.drop([0])

    file_name_ = file_name.split("\\")
    PSMs_df['File_name'] = file_name_[len(file_name_)-1]
    PSMs_df["predicted_retention_time"] = np.nan
    PSMs_df["tr"] = PSMs_df["retention_time"]
    PSMs_df["seq"] = PSMs_df["sequence"]
    PSMs_df["index"] = PSMs_df['spectra_ref']
    PSMs_df["localization_position"] = PSMs_df['opt_global_NuXL:best_localization_position']
    PSMs_df["localization_seq"] = PSMs_df["opt_global_NuXL:best_localization"]
    PSMs_df["localization_score"] = PSMs_df["opt_global_NuXL:best_localization_score"]
    PSMs_df["localization_scores"] = PSMs_df["opt_global_NuXL:localization_scores"]

    sequence = list(PSMs_df["seq"])
    PSMs_df = PSMs_df.astype({"localization_position":'int'})
    loc_pos = list(PSMs_df["localization_position"])
    peptido_form = list(PSMs_df['opt_global_cv_MS:1000889_peptidoform_sequence'])
    PSMs_df["peptidoform"] =  PSMs_df['opt_global_cv_MS:1000889_peptidoform_sequence']
    peptido_form_final = []
    for j in peptido_form:
        pattern = r'[0-9]'
        new_string = re.sub(pattern, '', j)
        pattern1 = r'[[].]'
        new_string_filter = re.sub(pattern1, '', new_string)
        
        if "PhosphoUridine" in j:
            #print(j,new_string_filter.replace("(PhosphoUridine)", ''))
            peptido_form_final.append(new_string_filter.replace("(PhosphoUridine)", ''))
        else:
            #print(j,new_string_filter)
            peptido_form_final.append(new_string_filter)

    PSMs_df['opt_global_cv_MS:1000889_peptidoform_sequence'] = peptido_form_final

    
    loc_Amino = []
    for i in range(len(loc_pos)):
        pos = loc_pos[i] 
        seq = sequence[i]
        AA = ""
        if (pos!= -1):
            AA = seq[pos]
        
        else:
            AA = "-"
        
        loc_Amino.append(AA)
    PSMs_df["loc_AA"] = loc_Amino 

    PSMs_df['Seq_modification'] = PSMs_df['sequence'] + '|' + PSMs_df['opt_global_NuXL:NA']
    PSMs_df['NA_mod'] = PSMs_df['opt_global_NuXL:NA']
    print(full_file_name, ">>>> PSMs_df: ", PSMs_df.shape)

    return PSMs_df #Dataframes of Return Description, Proteins and PSMs 
```

```python
#Take the most abundant peptide
def select_most_abundant(data_df):
    data_df["seq_RNA_mod"] = data_df["seq"] +"|"+ data_df["modifications"]
    data_df= data_df.sort_values(by=["seq_RNA_mod"])
    unique_seq = list(data_df["seq_RNA_mod"].unique())
    new_dataframe = []
    for i in unique_seq:
        x = data_df.loc[data_df['seq_RNA_mod'] == i]
        x_f = x[x["peptide_abundance_study_variable[1]"] == max(list(x["peptide_abundance_study_variable[1]"]))]
        new_dataframe.append(x_f.iloc[0].tolist())

    new_data_df = pd.DataFrame(new_dataframe)
    new_data_df.columns = data_df.columns
    return new_data_df
    
```
